## 問200（総合復習：クラスの定義）【最終問題】

円を表すクラス `Circle` を定義する。
初期化時に半径を受け取り、面積を返す `area` メソッドを持つ。
インスタンス自身を指す第1引数を空欄に入れよ。

```
class Circle:
    def __init__(___, r):
        self.r = r
    def area(self):
        return 3.14 * self.r ** 2

c = Circle(2)
print(c.area())
```
