### 問130
独自イテレータクラスで「次の要素を返す」ために実装する特殊メソッドを選べ。

- A. `__call__`
- B. `__next__`
- C. `__len__`
- D. `__get__`

```python
class Count:
    def __init__(self):
        self.n = 0
    def __iter__(self):
        return self
    def ___(self):
        self.n += 1
        if self.n > 2:
            raise StopIteration
        return self.n

for v in Count():
    print(v)
```
