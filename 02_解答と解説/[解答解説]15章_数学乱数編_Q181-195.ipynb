### 問181 の解答と解説

**正解： B) 5.0**

`math.sqrt(x)` は x の平方根を浮動小数点数(float)で返します。整数の平方根であっても常に float 型になるため、25 を渡しても 5(int) ではなく 5.0(float) になります。

```python
import math
print(math.sqrt(25))   # 5.0
```

### 問182 の解答と解説

**正解： B) 3 / -4**

`math.floor(x)` はxを超えない最大の整数を返します。正の数では単純な切り捨てですが、負の数では「よりマイナス側」に丸められる点に注意しましょう。-3.1 は -4 になります（-3 ではありません）。

```python
import math
print(math.floor(3.9))    # 3
print(math.floor(-3.1))   # -4
```

### 問183 の解答と解説

**正解： A) 4 / -3**

`math.ceil(x)` はx以上の最小の整数を返します。`math.floor`とは逆方向の丸めで、負の数では0に近い方向（この場合 -3）に丸められます。

```python
import math
print(math.ceil(3.1))    # 4
print(math.ceil(-3.9))   # -3
```

### 問184 の解答と解説

**正解： B) 8.0 / 8（float / int）**

`math.pow(x, y)` は常に float 型の結果を返しますが、組み込みのべき乗演算子 `**` は int 同士なら int を返します。両者は結果の値は同じでも型が異なる点に注意しましょう。

```python
import math
print(math.pow(2, 3))   # 8.0
print(2 ** 3)             # 8
```

### 問185 の解答と解説

**正解： C) 3 / -3**

`math.trunc(x)` は小数部を切り捨て、常に0に近い方向へ丸めます。`math.floor`が負の数でよりマイナス側に丸めるのに対し、`math.trunc`は符号を保ったまま整数部分だけを取り出します。

```python
import math
print(math.trunc(3.9))    # 3
print(math.trunc(-3.9))   # -3
```

### 問186 の解答と解説

**正解： A) 5.0 / 5（float / int）**

`math.fabs(x)` は常に float 型の絶対値を返しますが、組み込みの `abs()` は引数の型に応じてint ならint、float ならfloatを返します。数値全般に使える `abs()` のほうが一般的に使われます。

```python
import math
print(math.fabs(-5))   # 5.0
print(abs(-5))          # 5
```

### 問187 の解答と解説

**正解： C) 6**

`math.gcd(a, b)` はaとbの最大公約数を返します。12の約数は1,2,3,4,6,12、18の約数は1,2,3,6,9,18で、共通する最大の約数は6です。Python 3.9以降は複数の引数もまとめて渡せます。

```python
import math
print(math.gcd(12, 18))   # 6
```

### 問188 の解答と解説

**正解： B) 5.0**

`math.hypot(x, y)` は `sqrt(x**2 + y**2)` と同じ計算をして原点からの距離（ユークリッドノルム）を返します。3と4を辺とする直角三角形の斜辺は、3-4-5の直角三角形の性質から5.0になります。

```python
import math
print(math.hypot(3, 4))   # 5.0
```

### 問189 の解答と解説

**正解： A) 3.0**

`math.log(x, base)` はbaseを底とするxの対数を返します。2を底とする対数で 2**3 = 8 なので `math.log(8, 2)` は 3.0 です。底を省略すると自然対数(eを底とする対数)になります。

```python
import math
print(math.log(8, 2))   # 3.0
```

### 問190 の解答と解説

**正解： A) 0以上1未満のfloat**

`random.random()` は引数を取らず、0.0以上1.0未満のfloat型の乱数を返します。特定の範囲の整数が欲しい場合は `random.randint()` などを使います。

```python
import random
x = random.random()
print(0 <= x < 1)   # True
```

### 問191 の解答と解説

**正解： B) 1以上6以下の整数**

`random.randint(a, b)` はa以上b以下（両端を含む）の整数をランダムに返します。サイコロの目のように「1以上6以下」の乱数が欲しいときは `random.randint(1, 6)` のように使います。

```python
import random
n = random.randint(1, 6)
print(1 <= n <= 6)   # True
```

### 問192 の解答と解説

**正解： B) 1.0以上10.0以下のfloat**

`random.uniform(a, b)` はa以上b以下のfloat型の乱数を返します。`random.randint()` が整数専用なのに対し、`random.uniform()` は小数を含む連続的な範囲の乱数を生成したいときに使います。

```python
import random
x = random.uniform(1.0, 10.0)
print(1.0 <= x <= 10.0)   # True
```

### 問193 の解答と解説

**正解： A) リストの中からランダムに選ばれた1つの要素**

`random.choice(seq)` はシーケンスの中からランダムに1つの要素を選んで返します。空のシーケンスを渡すと `IndexError` になる点に注意しましょう。

```python
import random
fruits = ["apple", "banana", "cherry"]
print(random.choice(fruits) in fruits)   # True
```

### 問194 の解答と解説

**正解： A) nums の要素の並び順がその場で(in-place)シャッフルされる**

`random.shuffle(x)` はリストなどのミュータブルなシーケンスをその場で(in-place)シャッフルし、戻り値は `None` です。新しいリストを返すわけではないので、`nums = random.shuffle(nums)` のように書くと nums が None になってしまう典型的なミスに注意しましょう。

```python
import random
nums = [1, 2, 3, 4, 5]
result = random.shuffle(nums)
print(result)         # None
print(sorted(nums))   # [1, 2, 3, 4, 5]
```

### 問195 の解答と解説

**正解： A) 3 / 3（重複なしで3個選ばれる）**

`random.sample(population, k)` は母集団から重複させずにk個の要素をランダムに選んで新しいリストを返します。`random.choice()` を複数回呼び出す方法とは異なり、同じ要素が2回選ばれることはありません。

```python
import random
nums = [1, 2, 3, 4, 5]
result = random.sample(nums, 3)
print(len(result))        # 3
print(len(set(result)))   # 3(重複なし)
```
