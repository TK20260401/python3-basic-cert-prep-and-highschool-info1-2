### 問31 の解答と解説

**正解： B) [1, 2, 3, 4]**

`list.append(x)` はリストの末尾に要素を1つ追加します（元のリストを直接変更）。戻り値は `None` なので `nums = nums.append(4)` と書くと中身が消える点に注意。複数要素をまとめて追加したいときは `extend()` を使います。

```python
nums = [1, 2, 3]
nums.append(4)   # 末尾に追加
print(nums)   # [1, 2, 3, 4]
```

### 問32 の解答と解説

**正解： B) ['b', 'c', 'd']**

スライス `seq[start:stop]` は start を含み stop を含みません。`letters[1:4]` はインデックス1,2,3 の要素、すなわち 'b','c','d' を取り出します。start を省くと先頭から、stop を省くと末尾まで、という点も押さえましょう。

```python
letters = ['a', 'b', 'c', 'd', 'e']
print(letters[1:4])   # ['b', 'c', 'd']
```

### 問33 の解答と解説

**正解： B) 40 30**

負のインデックスは末尾から数えます。`-1` が最後の要素（40）、`-2` がその1つ前（30）です。長さを知らなくても末尾側の要素にアクセスできるので便利です。

```python
data = [10, 20, 30, 40]
print(data[-1], data[-2])   # 40 30
```

### 問34 の解答と解説

**正解： C) [0, 1, 4, 9, 16]**

リスト内包表記 `[式 for 変数 in 反復可能オブジェクト]` は、for ループを1行で書いてリストを生成します。`range(5)` は 0〜4 なので、各値の2乗は 0,1,4,9,16 になります。可読性を保てる範囲で使うのがコツです。

```python
squares = [x**2 for x in range(5)]
print(squares)   # [0, 1, 4, 9, 16]
```

### 問35 の解答と解説

**正解： B) [1, 2, 3] [3, 1, 2]**

組み込み関数 `sorted()` は並べ替えた「新しいリスト」を返し、元のリストは変更しません。一方、リストのメソッド `list.sort()` は元のリスト自体を並べ替え、戻り値は `None` です。この違いが頻出ポイントです。

```python
nums = [3, 1, 2]
result = sorted(nums)   # 新しいリストを返す
print(result, nums)   # [1, 2, 3] [3, 1, 2]
```

### 問36 の解答と解説

**正解： A) 10**

タプルは `( )` で作る「変更できない（イミュータブル）シーケンス」です。要素の参照はリストと同じく `[インデックス]` で行い、`point[0]` は最初の要素 10 を返します。`point[0] = 5` のような代入はエラーになります。

```python
point = (10, 20)
print(point[0])   # 10
```

### 問37 の解答と解説

**正解： C) 3**

`a, b = (1, 2)` のように、タプルの各要素を複数の変数へ一度に代入することをアンパック代入といいます。ここでは a=1, b=2 となり、a+b は 3。`a, b = b, a` で値の入れ替えができるのも同じ仕組みです。

```python
a, b = (1, 2)   # アンパック代入
print(a + b)   # 3
```

### 問38 の解答と解説

**正解： B) 3**

`set()` は重複を持たない集合を作ります。`[1,2,2,3,3,3]` から重複を除くと {1,2,3} の3要素になり、`len()` は 3 を返します。集合は要素の順序を保持しない点にも注意しましょう。

```python
nums = [1, 2, 2, 3, 3, 3]
unique = set(nums)   # 重複を除去
print(len(unique))   # 3
```

### 問39 の解答と解説

**正解： B) {2, 3}**

`&` は集合の積（共通部分）を求める演算子です。a と b の両方に含まれる要素 2 と 3 が結果になります。他に和集合は `|`、差集合は `-`、対称差は `^` です。

```python
a = {1, 2, 3}
b = {2, 3, 4}
print(a & b)   # {2, 3}
```

### 問40 の解答と解説

**正解： B) 90**

辞書は `辞書[キー]` で値を取得します。`scores["eng"]` はキー "eng" に対応する値 90 を返します。存在しないキーを指定すると `KeyError` になる点に注意（回避には `get()` が便利）。

```python
scores = {"math": 80, "eng": 90}
print(scores["eng"])   # 90
```

### 問41 の解答と解説

**正解： B) {"a": 1, "b": 2}**

辞書に新しいキーと値を追加するには `辞書[新しいキー] = 値` と書きます。既存のキーを指定すると値が上書きされます。`d["b"] = 2` でキー "b"・値 2 が加わり、{'a':1, 'b':2} になります。

```python
d = {"a": 1}
d["b"] = 2   # キーと値を追加
print(d)   # {'a': 1, 'b': 2}
```

### 問42 の解答と解説

**正解： B) 0**

`dict.get(キー, 既定値)` はキーがあればその値を、無ければ既定値を返します。`d.get("y", 0)` はキー "y" が無いので 0 を返します。第2引数を省くと `None` が返り、`[]` と違って `KeyError` にならないのが利点です。

```python
d = {"x": 1}
print(d.get("y", 0))   # 0（キーが無いので既定値）
```

### 問43 の解答と解説

**正解： C) a 1 と b 2（キーと値の組）**

`dict.items()` はキーと値の組（タプル）を順に返します。`for k, v in d.items():` でキーを k、値を v に受け取れます。キーだけなら `keys()`、値だけなら `values()` を使います。

```python
d = {"a": 1, "b": 2}
for k, v in d.items():   # キーと値の組
    print(k, v)   # a 1 / b 2
```

### 問44 の解答と解説

**正解： A) True**

辞書に対する `in` 演算子は「キー」の存在を判定します。"apple" はキーとして存在するので `True`。値を対象に調べたいときは `"apple" in d.values()` のようにします。

```python
d = {"apple": 100, "banana": 200}
print("apple" in d)   # True（キーの存在確認）
```

### 問45 の解答と解説

**正解： B) Ben**

ネストした構造は外側から順にたどります。`data["users"]` はリスト、その `[1]` は2番目の辞書 {"name": "Ben"}、さらに `["name"]` で "Ben" を取得します。辞書とリストを組み合わせた表現（JSONなど）でよく使う操作です。

```python
data = {"users": [{"name": "Aoi"}, {"name": "Ben"}]}
print(data["users"][1]["name"])   # Ben
```
