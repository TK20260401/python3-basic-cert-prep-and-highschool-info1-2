### 問136 解答
**正解: A. `search`**

`re.search(pattern, string)` は文字列全体を走査し、パターンに最初にマッチする部分を探します。マッチした場合は `Match` オブジェクトを返し、`.group()` で一致した文字列を取得できます。出力は `3.12` です。

```python
import re
text = "Python 3.12 is great"
m = re.search(r"\d+\.\d+", text)
print(m.group())
```

### 問137 解答
**正解: A. `findall`**

`re.findall(pattern, string)` はパターンに一致する部分をすべて探し、リストとして返します。出力は `['100', '200', '300']` です。

```python
import re
text = "apple:100, banana:200, cherry:300"
nums = re.findall(r"\d+", text)
print(nums)
```

### 問138 解答
**正解: A. `sub`**

`re.sub(pattern, repl, string)` はパターンに一致した部分を置換文字列に置き換えた新しい文字列を返します。出力は `2026/07/21` です。

```python
import re
text = "2026-07-21"
result = re.sub(r"-", "/", text)
print(result)
```

### 問139 解答
**正解: A. `split`**

`re.split(pattern, string)` はパターンに一致する部分を区切りとして文字列を分割し、リストで返します。出力は `['a', 'b', 'c', 'd']` です。

```python
import re
text = "a1b22c333d"
parts = re.split(r"\d+", text)
print(parts)
```

### 問140 解答
**正解: A. `compile`**

`re.compile(pattern)` は正規表現パターンを事前にコンパイルし、`Pattern` オブジェクトとして返します。同じパターンを繰り返し使う場合に効率的で、`.findall()` などのメソッドをそのまま呼び出せます。出力は `['12', '34']` です。

```python
import re
pattern = re.compile(r"\d+")
print(pattern.findall("A12 B34"))
```

### 問141 解答
**正解: A. `Counter`**

`collections.Counter(iterable)` は各要素の出現回数を数え、辞書のようなオブジェクトとして保持します。`counts["a"]` は3を返します。

```python
from collections import Counter
data = ["a", "b", "a", "c", "b", "a"]
counts = Counter(data)
print(counts["a"])
```

### 問142 解答
**正解: A. `defaultdict`**

`collections.defaultdict(default_factory)` は存在しないキーへアクセスした際に、`default_factory` を呼び出して初期値を自動生成します。出力は `['apple']` です。

```python
from collections import defaultdict
d = defaultdict(list)
d["fruits"].append("apple")
print(d["fruits"])
```

### 問143 解答
**正解: A. `namedtuple`**

`collections.namedtuple(typename, field_names)` はフィールド名で属性アクセスできるタプルのサブクラスを生成します。出力は `3 4` です。

```python
from collections import namedtuple
Point = namedtuple("Point", ["x", "y"])
p = Point(3, 4)
print(p.x, p.y)
```

### 問144 解答
**正解: A. `deque`**

`collections.deque` は両端キューで、`appendleft`/`append` により先頭・末尾への追加をO(1)で行えます。出力は `[0, 1, 2, 3, 4]` です。

```python
from collections import deque
dq = deque([1, 2, 3])
dq.appendleft(0)
dq.append(4)
print(list(dq))
```

### 問145 解答
**正解: A. `chain`**

`itertools.chain(*iterables)` は複数のiterableを順につなげた1つのイテレータを返します。出力は `[1, 2, 3, 4, 5]` です。

```python
from itertools import chain
result = list(chain([1, 2], [3, 4], [5]))
print(result)
```

### 問146 解答
**正解: A. `product`**

`itertools.product(*iterables)` は複数のiterableのデカルト積（すべての組み合わせ）をタプルとして生成します。出力は `[(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')]` です。

```python
from itertools import product
result = list(product([1, 2], ["a", "b"]))
print(result)
```

### 問147 解答
**正解: A. `permutations`**

`itertools.permutations(iterable, r)` は要素の並び順を区別した順列を生成します。3要素から2つ選ぶ順列は6通りのため、出力は `6` です。

```python
from itertools import permutations
result = list(permutations([1, 2, 3], 2))
print(len(result))
```

### 問148 解答
**正解: A. `combinations`**

`itertools.combinations(iterable, r)` は要素の並び順を区別しない組み合わせを生成します。3要素から2つ選ぶ組み合わせは3通りのため、出力は `3` です。

```python
from itertools import combinations
result = list(combinations([1, 2, 3], 2))
print(len(result))
```

### 問149 解答
**正解: A. `reduce`**

`functools.reduce(function, iterable)` は要素を左から順に累積的に演算し、最終的に1つの値にまとめます。`1*2*3*4=24` のため出力は `24` です。

```python
from functools import reduce
result = reduce(lambda a, b: a * b, [1, 2, 3, 4])
print(result)
```

### 問150 解答
**正解: A. `lru_cache`**

`functools.lru_cache()` デコレータを付けると、同じ引数での呼び出し結果をキャッシュし、2回目以降は関数本体を再実行せずキャッシュ値を返します。そのため `"計算中: 4"` の表示は1回だけで、`16` が2回出力されます。

```python
from functools import lru_cache

@lru_cache()
def square(n):
    print("計算中:", n)
    return n * n

print(square(4))
print(square(4))
```
