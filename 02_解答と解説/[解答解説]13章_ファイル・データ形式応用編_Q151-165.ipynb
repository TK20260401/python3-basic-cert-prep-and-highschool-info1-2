### 問151 解答
**正解: A. `/`**

pathlibでは `/` 演算子がオーバーロードされており、Pathと文字列（またはPath同士）を自然な記述で結合できます。出力は `data/file.txt` です。

```python
from pathlib import Path
p = Path("data") / "file.txt"
print(p)
```

### 問153 解答
**正解: A. `mkdir`**

`Path.mkdir(parents=True, exist_ok=True)` では、途中の親ディレクトリがなくても自動で作成し、すでに存在していてもエラーになりません。出力は `True` です。

### 問154 解答
**正解: A. `glob`**

`Path.glob(pattern)` はワイルドカードパターンに一致するパスをイテレータとして返します。`*.txt` は `.txt` 拡張子のファイルに一致します。出力は `['work/a.txt', 'work/b.txt']` です。

```python
from pathlib import Path
folder = Path("work")
folder.mkdir(exist_ok=True)
(folder / "a.txt").write_text("A")
(folder / "b.txt").write_text("B")
(folder / "c.csv").write_text("C")
txt_files = sorted(str(f) for f in folder.glob("*.txt"))
print(txt_files)
```

### 問155 解答
**正解: A. `read_text`**

`Path.read_text()` はファイルを開いて内容を文字列として読み込み、自動的にクローズします。`write_text()` と対になるメソッドです。出力は `Hello, Python!` です。

```python
from pathlib import Path
p = Path("work/note.txt")
p.write_text("Hello, Python!")
print(p.read_text())
```

### 問156 解答
**正解: A. `dumps`**

`json.dumps(obj)` はオブジェクトをJSON形式の文字列に変換します（末尾の s は stringの意味）。ファイルに直接書き込む場合は `dump` を使います。出力は `{"name": "Alice", "age": 20}` です。

```python
import json
data = {"name": "Alice", "age": 20}
s = json.dumps(data)
print(s)
```

### 問157 解答
**正解: A. `loads`**

`json.loads(s)` はJSON形式の文字列を解析してPythonの辞書（やリストなど）に変換します。`1 + 2` なので出力は `3` です。

```python
import json
s = '{"x": 1, "y": 2}'
data = json.loads(s)
print(data["x"] + data["y"])
```

### 問158 解答
**正解: A. `dump`**

`json.dump(obj, f, ensure_ascii=False)` はオブジェクトをJSONとしてファイルに直接書き辺みます。`ensure_ascii=False` にすることで日本語が `\uXXXX` ではなくそのまま出力されます。出力は `{"名前": "太郎", "点数": 90}` です。

```python
import json
data = {"名前": "太郎", "点数": 90}
with open("work/data.json", "w", encoding="utf-8") as f:
 json.dump(data, f, ensure_ascii=False)
print(open("work/data.json", encoding="utf-8").read())
```

### 問159 解答
**正解: A. `load`**

`json.load(f)` はファイルオブジェクトからJSONを読み込み、Pythonオブジェクト（辞書など）に変換します。出力は `90` です。

```python
import json
with open("work/data.json", encoding="utf-8") as f:
 loaded = json.load(f)
print(loaded["点数"])
```

### 問160 解答
**正解: A. `writer`**

`csv.writer(f)` はリストをCSVの1行として書き込むライターオブジェクトを作成します。`writerow()` で1行ずつ追加します。出力は `['name,score', 'Bob,85']` です。

```python
import csv
with open("work/sample.csv", "w", newline="", encoding="utf-8") as f:
 w = csv.writer(f)
 w.writerow(["name", "score"])
 w.writerow(["Bob", 85])

with open("work/sample.csv", encoding="utf-8") as f:
 print(f.read().splitlines())
```

### 問161 解答
**正解: A. `reader`**

`csv.reader(f)` はCSVファイルを、1行ごとにリストとして返すイテレータを作成します。出力は `[['name', 'score'], ['Bob', '85']]` です。

```python
import csv
with open("work/sample.csv", encoding="utf-8") as f:
 r = csv.reader(f)
 rows = list(r)
print(rows)
```

### 問162 解答
**正解: A. `DictWriter`**

`csv.DictWriter(f, fieldnames=...)` は辞書をキーに応じた列に展開してCSVに書き込みます。`writeheader()` でヘッダー行を書きます。出力は `['name,score', 'Carol,92']` です。

```python
import csv
fieldnames = ["name", "score"]
with open("work/dict_sample.csv", "w", newline="", encoding="utf-8") as f:
 w = csv.DictWriter(f, fieldnames=fieldnames)
 w.writeheader()
 w.writerow({"name": "Carol", "score": 92})

with open("work/dict_sample.csv", encoding="utf-8") as f:
 print(f.read().splitlines())
```

### 問163 解答
**正解: A. `DictReader`**

`csv.DictReader(f)` はヘッダー行をキーとして各行を辞書として読み込みます。出力は `Carol` です。

```python
import csv
with open("work/dict_sample.csv", encoding="utf-8") as f:
 r = csv.DictReader(f)
 rows = list(r)
print(rows[0]["name"])
```

### 問164 解答
**正解: A. `copy`**

`shutil.copy(src, dst)` はファイルの内容とパーミッションをコピーし、コピー先のパスを返します。元ファイルは残ります。出力は `data` です。

```python
import shutil
from pathlib import Path
Path("work/src.txt").write_text("data")
shutil.copy("work/src.txt", "work/dst.txt")
print(Path("work/dst.txt").read_text())
```

### 問165 解答
**正解: A. `move`**

`shutil.move(src, dst)` はファイルやディレクトリを移動（リネームも含む）します。移動後は元のパスには存在しなくなります。出力は `True False` です。

```python
import shutil
from pathlib import Path
Path("work/moved_src.txt").write_text("moved")
shutil.move("work/moved_src.txt", "work/moved_dst.txt")
print(Path("work/moved_dst.txt").exists(), Path("work/moved_src.txt").exists())
```
