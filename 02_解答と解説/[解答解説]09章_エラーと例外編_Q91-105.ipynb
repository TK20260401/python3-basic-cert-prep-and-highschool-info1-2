# 第9章 エラーと例外編（Q91-105）解答と解説

各問題の正解と日本語解説。先生用（配布用ではありません）。

### 問91 解答：B. `except`
Python では `try` ブロックで例外を監視し、`except 例外型:` で捕捉して処理します（`catch` ではありません）。

```python
try:
    x = 1 / 0
except ZeroDivisionError:
    print('エラーが発生しました')
```

### 問92 解答：C. `as`
`except 例外型 as 変数:` の形で、捕捉した例外オブジェクトを変数に束縛できます。`print(e)` でエラーメッセージを表示できます。

```python
try:
    int('abc')
except ValueError as e:
    print('変換失敗:', e)
```

### 問93 解答：A. `finally`
`finally` 節は例外が起きても起きなくても、`try` 文を抜ける前に必ず実行されます。ファイルのクローズなど後始末に使います。

```python
try:
    f = open('data.txt')
except FileNotFoundError:
    print('ファイルなし')
finally:
    print('処理終了')
```

### 問94 解答：B. `else`
`try` 文の `else` 節は、`try` ブロックで例外が発生しなかった場合にのみ実行されます。`except` の後、`finally` の前に置きます。

```python
try:
    n = int('10')
except ValueError:
    print('変換失敗')
else:
    print('変換成功:', n)
```

### 問95 解答：B. `raise`
`raise 例外(メッセージ)` で例外を送出します。Python では `throw` ではなく `raise` を使います。

```python
def check(age):
    if age < 0:
        raise ValueError('年齢が負です')
    return age
```

### 問96 解答：A. `except KeyError:`
`try` に対して `except` を複数並べると、発生した例外の型に一致する最初の `except` が実行されます。存在しないキー参照は `KeyError` です。

```python
d = {}
try:
    d['x']
except IndexError:
    print('範囲外')
except KeyError:
    print('キーがありません')
```

### 問97 解答：C. `Exception`
`Exception` は組み込み例外の基底クラスで、`except Exception:` とすると多くの例外をまとめて捕捉できます（`SystemExit` などは `BaseException` 側）。

```python
try:
    risky()
except Exception as e:
    print('何らかのエラー:', e)
```

### 問98 解答：C. `KeyError`
辞書に無いキーを `d[key]` で参照すると `KeyError` が発生します。リストの範囲外は `IndexError` です。

```python
d = {'a': 1}
try:
    print(d['b'])
except KeyError:
    print('そのキーはありません')
```

### 問99 解答：B. `IndexError`
シーケンス（リストやタプル）で存在しない位置を参照すると `IndexError` が発生します。

```python
lst = [1, 2, 3]
try:
    print(lst[10])
except IndexError:
    print('範囲外です')
```

### 問100 解答：C. `TypeError`
演算子や関数に不適切な「型」の値を渡すと `TypeError` が発生します。値は正しいが内容が不適切な場合は `ValueError` です。

```python
try:
    x = 'a' + 1
except TypeError:
    print('型が合いません')
```

### 問101 解答：B. `ValueError`
引数の型は正しいものの値が不適切な場合に `ValueError` が発生します。`int('abc')` は代表例です。

```python
try:
    n = int('abc')
except ValueError:
    print('数値に変換できません')
```

### 問102 解答：C. `Exception`
独自例外は通常 `Exception`（またはそのサブクラス）を継承して定義します。これにより `except Exception` でまとめて捕捉できます。

```python
class MyError(Exception):
    pass

raise MyError('独自エラー')
```

### 問103 解答：A. `raise`
`except` 節の中で引数なしの `raise` を書くと、現在処理中の例外をそのまま再送出できます。ログ記録後に上位へ伝播させる用途で使います。

```python
try:
    risky()
except Exception:
    print('記録した')
    raise
```

### 問104 解答：B. `assert`
`assert 条件, メッセージ` は条件が偽のとき `AssertionError` を送出します。デバッグや前提条件の確認に使います。

```python
def sqrt(x):
    assert x >= 0, 'x は非負である必要があります'
    return x ** 0.5
```

### 問105 解答：B. `with`
`with` 文はコンテキストマネージャを使い、ブロックを抜けるとき（例外時も含め）に自動で後始末（ファイルのクローズなど）を行います。

```python
with open('data.txt') as f:
    data = f.read()
```
