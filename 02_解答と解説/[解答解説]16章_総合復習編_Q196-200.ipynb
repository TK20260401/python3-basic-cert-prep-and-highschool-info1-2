## 解答196

**正解:** `n ** 2`

リスト内包表記 `[式 for 変数 in 反復可能 if 条件]` は、
条件を満たす要素だけに式を適用して新しいリストを作る。
ここでは偶数(`n % 2 == 0`)を2乗している。

```
nums = [1, 2, 3, 4, 5, 6]
result = [n ** 2 for n in nums if n % 2 == 0]
print(result)
```

## 解答197

**正解:** `items`

`dict.items()` は (キー, 値) のタプルを順に返す。
`for k, v in scores.items():` のように
キーと値を同時に受け取れる。

```
scores = {'国語': 80, '数学': 95}
for k, v in scores.items():
    print(f'{k}: {v}')
```

## 解答198

**正解:** `try`

`try:` ブロックで例外が発生する可能性のある処理を囲み、
`except ValueError:` で変換失敗時の処理を書く。
これにより 'abc' でもプログラムは止まらない。

```
for s in ['10', 'abc', '5']:
    try:
        print(int(s))
    except ValueError:
        print('変換できません')
```

## 解答199

**正解:** `='ゲスト'`

仮引数に `name='ゲスト'` と書くと、
呼び出し時に引数を省略した場合の既定値になる。
`greet()` は 'ゲスト'、`greet('田中')` は '田中' を使う。

```
def greet(name='ゲスト'):
    return f'こんにちは、{name}さん'

print(greet())
print(greet('田中'))
```

## 解答200【最終問題・全200問完結】

**正解:** `self`

メソッドの第1引数 `self` はインスタンス自身を指す慣例名。
`__init__` で `self.r = r` と属性を設定し、
`area` で半径から面積を計算している。

🎉 これで全200問が完成しました。お疲れさまでした！

```
class Circle:
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r ** 2

c = Circle(2)
print(c.area())
```
