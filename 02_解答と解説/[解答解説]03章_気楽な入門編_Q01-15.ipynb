### 問1 の解答と解説

**正解： A) 5.666666666666667 / 5 / 2**

Python 3 では割り算 `/` は常に float を返すため `17 / 3` は 5.666… になります。`//` は切り捨て除算で商 `5`、`%` は余り `2` を返します。（商 × 除数 + 余り = 元の数：5×3+2=17）

```python
print(17 / 3)    # 5.666666666666667
print(17 // 3)   # 5
print(17 % 3)    # 2
```

### 問2 の解答と解説

**正解： B) 1024 / 25**

`**` はべき乗演算子です。`2 ** 10` は 1024、`5 ** 2` は 25。なお `-3 ** 2` は `-(3 ** 2)` すなわち `-9` になる点も応用として押さえましょう。

```python
print(2 ** 10)   # 1024
print(5 ** 2)    # 25
```

### 問3 の解答と解説

**正解： B) 1行目は改行やタブが解釈され、2行目は `C:\name\test`**

通常の文字列では `\n` は改行、`\t` はタブとして解釈されます。`r"..."` の raw文字列はバックスラッシュを特殊文字として扱わず、書いたとおりに出力します。

### 問4 の解答と解説

**正解： B) 先頭に空行が入らず、`Usage:` の行から表示される**

### 問5 の解答と解説

**正解： A) P / n**

インデックスは 0 から始まるため `word[0]` は "P"。負のインデックスは末尾から数え `-1` が最後の文字を指すので `word[-1]` は "n" です。

```python
word = "Python"
print(word[0])    # P
print(word[-1])   # n
```

### 問6 の解答と解説

**正解： A) Pyt / thon**

`start` を省略すると 0 から、`stop` を省略すると末尾までを意味します。`word[:3]` は "Pyt"、`word[2:]` は "thon"。`word[:i] + word[i:]` が元の文字列に等しくなります。

```python
word = "Python"
print(word[:3])   # Pyt
print(word[2:])   # thon
```

### 問7 の解答と解説

**正解： B) on / Pto**

スライスは範囲外を指定してもエラーにならず、存在する範囲に調整されます。`word[4:100]` は "on"。`[::2]` はステップ2で P(0), t(2), o(4) の "Pto" になります。

```python
word = "Python"
print(word[4:100])   # on
print(word[::2])     # Pto
```

### 問8 の解答と解説

**正解： C) TypeError が発生する**

文字列はイミュータブル（変更不可）で、要素を直接書き換えると `TypeError` になります。変更したい場合はスライスや連結で新しい文字列を作ります。

```python
word = "Python"
new_word = "J" + word[1:]
print(new_word)   # Jython
```

### 問9 の解答と解説

**正解： A) 20 / 4**

`len()` は文字列なら文字数、リストなら要素数を返します。"supercalifragilistic" は20文字、`[10,20,30,40]` は4要素です。

```python
s = "supercalifragilistic"
nums = [10, 20, 30, 40]
print(len(s))      # 20
print(len(nums))   # 4
```

### 問10 の解答と解説

**正解： A) 1 / 25 / [9, 16, 25]**

`squares[0]` は先頭の 1、`squares[-1]` は末尾の 25、`squares[2:]` はインデックス2から末尾までの新しいリスト `[9, 16, 25]` です。単体インデックスは要素そのもの、スライスはリストを返す点が違いです。

```python
squares = [1, 4, 9, 16, 25]
print(squares[0])    # 1
print(squares[-1])   # 25
print(squares[2:])   # [9, 16, 25]
```

### 問12 の解答と解説

**正解： A) ['a','b','C','D','E','f'] / ['a','b','f']**

スライスにリストを代入すると範囲をまとめて置き換えられます。さらに空リスト `[]` を代入するとその範囲が削除され、`['a', 'b', 'f']` になります。

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f']
letters[2:5] = ['C', 'D', 'E']
print(letters)   # ['a', 'b', 'C', 'D', 'E', 'f']
letters[2:5] = []
print(letters)   # ['a', 'b', 'f']
```

### 問13 の解答と解説

**正解： B) [1, 4, 9, 16, 25]**

`append()` はリストの末尾に1つの要素を追加します。`5 ** 2` は 25。`append()` はその場でリストを変更し、戻り値は `None` である点にも注意しましょう。

```python
squares = [1, 4, 9]
squares.append(16)
squares.append(5 ** 2)
print(squares)   # [1, 4, 9, 16, 25]
```

### 問14 の解答と解説

**正解： A) ['a', 'b', 'c'] / b**

`x = [a, n]` は `[['a','b','c'], [1,2,3]]`。`x[0]` は1つ目のリスト全体、`x[0][1]` はそのリストのインデックス1なので `'b'`。`[外側][内側]` の順でアクセスします。

```python
a = ['a', 'b', 'c']
n = [1, 2, 3]
x = [a, n]
print(x[0])      # ['a', 'b', 'c']
print(x[0][1])   # b
```

### 問15 の解答と解説

**正解： A) 12.5625 → 12.56**

`price * tax` は `100.50 * 0.125 = 12.5625`。対話モードでは直前に表示された値が特殊変数 `_` に入るため、`round(_, 2)` は 12.56 を返します。`_` は読み取り専用として使うのがマナーです。Colabのコードセルでは行の値が自動表示されないため、演習では `result` と `print()` で再現しています。

```python
tax = 12.5 / 100
price = 100.50
result = price * tax
print(result)              # 12.5625
print(round(result, 2))    # 12.56
```
