### 問106 解答：B. `class`
クラスは `class クラス名:` で定義します。慣習としてクラス名は先頭大文字（キャメルケース）で書きます。

```python
class Dog:
    pass

d = Dog()
```

### 問107 解答：B. `__init__`
`__init__` はインスタンス生成時に自動的に呼ばれる初期化メソッドです。第1引数の `self` を通じて属性を設定します。

```python
class Dog:
    def __init__(self, name):
        self.name = name

d = Dog('ポチ')
```

### 問108 解答：B. `self`
インスタンスメソッドの第1引数には、呼び出したインスタンス自身が自動で渡されます。慣習的に `self` という名前を使います。

```python
class Dog:
    def __init__(self, name):
        self.name = name
```

### 問109 解答：A. `self.name`
インスタンス属性はメソッド内で `self.属性名` として参照します。`self` を付けないと単なるローカル変数扱いになります。

```python
class Dog:
    def __init__(self, name):
        self.name = name
    def bark(self):
        return self.name + ' がワン'

print(Dog('ポチ').bark())
```

### 問110 解答：C. `def`
メソッドはクラスの中で `def` を使って定義します。第1引数に `self` を取る点が通常の関数と異なります。

```python
class Counter:
    def __init__(self):
        self.n = 0
    def increment(self):
        self.n += 1
```

### 問111 解答：B. `legs = 4`
メソッドの外側（クラス本体直下）に書いた変数は「クラス属性」となり、全インスタンスで共有されます。`クラス名.属性` で参照できます。

```python
class Dog:
    legs = 4
    def __init__(self, name):
        self.name = name

print(Dog.legs)
```

### 問112 解答：A. `class Dog(Animal):`
継承は `class 子クラス(親クラス):` の形で書きます。子クラスは親クラスの属性やメソッドを引き継ぎます。

```python
class Animal:
    pass

class Dog(Animal):
    pass
```

### 問113 解答：C. `super`
`super().__init__(...)` で親クラスの初期化メソッドを呼び出せます。親の初期化処理を再利用しつつ、子で追加の処理を書けます。

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
```

### 問114 解答：A. `def speak(self):`
子クラスで親と同じ名前のメソッドを定義すると、親のメソッドを上書き（オーバーライド）できます。呼び出すと子の定義が優先されます。

```python
class Animal:
    def speak(self):
        return '...'

class Dog(Animal):
    def speak(self):
        return 'ワン'
```

### 問115 解答：B. `__str__`
`__str__` は `print()` や `str()` で使われる「人間に読みやすい」文字列表現を返します。開発者向けの詳細表現は `__repr__` です。

```python
class Dog:
    def __init__(self, name):
        self.name = name
    def __str__(self):
        return f'Dog({self.name})'

print(Dog('ポチ'))
```

### 問116 解答：C. `Dog()`
Python では `クラス名()` を呼び出すとインスタンスが生成されます。`new` キーワードは不要です。

```python
class Dog:
    pass

d = Dog()
print(type(d))
```

### 問117 解答：B. `isinstance`
`isinstance(obj, クラス)` はオブジェクトがそのクラス（または子クラス）のインスタンスなら `True` を返します。クラス同士の関係判定は `issubclass` です。

```python
class Dog:
    pass

d = Dog()
print(isinstance(d, Dog))
```

### 問118 解答：C. `@classmethod`
`@classmethod` を付けたメソッドは第1引数にクラス自身（慣習的に `cls`）を受け取ります。クラス変数の操作やファクトリメソッドに使います。

```python
class Dog:
    count = 0
    @classmethod
    def create(cls):
        cls.count += 1
        return cls()
```

### 問119 解答：B. `@staticmethod`
`@staticmethod` を付けたメソッドは `self`／`cls` を受け取らず、クラスに関連づけた通常の関数として動作します。

```python
class MathUtil:
    @staticmethod
    def add(a, b):
        return a + b

print(MathUtil.add(2, 3))
```

### 問120 解答：A. `@property`
`@property` を付けると、メソッドを `()` なしで属性のように参照できます。計算で求まる値を属性風に見せるときに使います。

```python
class Circle:
    def __init__(self, r):
        self.r = r
    @property
    def area(self):
        return 3.14 * self.r ** 2

print(Circle(2).area)
```
