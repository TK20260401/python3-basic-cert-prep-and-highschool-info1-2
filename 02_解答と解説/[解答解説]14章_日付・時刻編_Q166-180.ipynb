### 問166 解答
**正解: A. `datetime.date`**

`datetime.date(year, month, day)` で日付（年・月・日）のみを表すオブジェクトを作成できます。出力は `2024-04-01` です。

```python
import datetime
d = datetime.date(2024, 4, 1)
print(d)
```

### 問167 解答
**正解: A. `.month`**

`datetime` オブジェクトは `.year` `.month` `.day` `.hour` `.minute` `.second` などの属性を持ちます。`dt.month` は `4` です。

```python
import datetime
dt = datetime.datetime(2024, 4, 1, 9, 30, 0)
print(dt.month)
```

### 問168 解答
**正解: A. `datetime.date.today()`**

`datetime.date.today()` は実行時点の日付（年・月・日）を `date` オブジェクトとして返します。時刻を含む現在日時が必要な場合は `datetime.datetime.now()` を使います。

```python
import datetime
today = datetime.date.today()
print(type(today))
```

### 問169 解答
**正解: A. `datetime.timedelta`**

`datetime.timedelta(days=7)` は「7日間」という期間を表すオブジェクトです。日付や日時に加算・減算することで、未来や過去の日付を求められます。結果は `2024-04-08` です。

```python
import datetime
d = datetime.date(2024, 4, 1)
future = d + datetime.timedelta(days=7)
print(future)
```

### 問170 解答
**正解: A. `-`**

`date` 同士を `-` で引き算すると `timedelta` オブジェクトが得られます。`.days` 属性で日数（整数）を取り出せます。結果は `9` です。

```python
import datetime
d1 = datetime.date(2024, 4, 10)
d2 = datetime.date(2024, 4, 1)
diff = d1 - d2
print(diff.days)
```

### 問171 解答
**正解: A. `.days`**

`timedelta` は `.days`（日数）、`.seconds`（日をまたいだ残り秒数）、`.microseconds` の3属性を持ちます。`delta.days` は `5` です。

```python
import datetime
delta = datetime.timedelta(days=5, hours=3)
print(delta.days)
```

### 問172 解答
**正解: A. `.strftime()`**

`.strftime(format)` はフォーマットコード（`%Y` `%m` `%d` `%H` `%M` など）を指定して、datetimeを任意の書式の文字列に変換します。出力は `2024/04/01 09:05` です。

```python
import datetime
dt = datetime.datetime(2024, 4, 1, 9, 5, 0)
print(dt.strftime("%Y/%m/%d %H:%M"))
```

### 問173 解答
**正解: A. `datetime.datetime.strptime`**

`strptime(文字列, フォーマット)` は文字列を指定書式で解析し、datetimeオブジェクトに変換します。`strftime` とは逆方向の変換です。結果は `2024-04-01 00:00:00` です。

```python
import datetime
s = "2024-04-01"
dt = datetime.datetime.strptime(s, "%Y-%m-%d")
print(dt)
```

### 問174 解答
**正解: A. `H`**

`%H` は24時間表記の時（00〜23）をゼロ埋め2桁で表します。`%M` は分、`%S` は秒です。出力は `09:05:03` です。

```python
import datetime
dt = datetime.datetime(2024, 4, 1, 9, 5, 3)
print(dt.strftime("%H:%M:%S"))
```

### 問175 解答
**正解: A. `d1 < d2`**

`date`（および`datetime`）オブジェクトは通常の比較演算子 `<` `>` `<=` `>=` `==` で大小・等価を比較できます。`d1` の方が過去なので `True` になります。

```python
import datetime
d1 = datetime.date(2024, 1, 1)
d2 = datetime.date(2024, 12, 31)
print(d1 < d2)
```

### 問176 解答
**正解: A. `time.time()`**

`time.time()` は1970年1月1日からの経過秒数（エポック秒）をfloatで返します。処理時間の計測などによく使われます。

```python
import time
t = time.time()
print(type(t))
```

### 問177 解答
**正解: A. `time.sleep()`**

`time.sleep(秒数)` はプログラムの実行を指定秒数だけ一時停止します。1秒待機した後の経過時間は1秒以上になるため `True` が出力されます。

```python
import time
start = time.time()
time.sleep(1)
elapsed = time.time() - start
print(elapsed >= 1)
```

### 問178 解答
**正解: A. `weeks=2`**

`timedelta` は `weeks` キーワード引数を受け付け、内部的に `days` に変換されます（1週間 = 7日）。`weeks=2` は14日後を意味し、結果は `2024-04-15` です。

```python
import datetime
d = datetime.date(2024, 4, 1)
future = d + datetime.timedelta(weeks=2)
print(future)
```

### 問179 解答
**正解: A. `0`**

`weekday()` は月曜日を `0`、火曜日を `1`、…、日曜日を `6` として返します（`isoweekday()` は月曜日が `1` になる点に注意）。2024年4月1日は月曜日なので `True` です。

```python
import datetime
d = datetime.date(2024, 4, 1)  # 2024年4月1日は月曜日
print(d.weekday() == 0)
```

### 問180 解答
**正解: A. `.isoformat()`**

`.isoformat()` はdatetimeを `YYYY-MM-DDTHH:MM:SS` 形式（ISO 8601）の文字列に変換します。出力は `2024-04-01T09:05:03` です。

```python
import datetime
dt = datetime.datetime(2024, 4, 1, 9, 5, 3)
print(dt.isoformat())
```
