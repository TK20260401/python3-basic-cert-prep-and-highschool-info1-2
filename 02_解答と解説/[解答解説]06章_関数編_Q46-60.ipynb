### 問46 の解答と解説

**正解： A) Hello, Python**

関数は `def 関数名(引数):` で定義します。`greet("Python")` と呼び出すと引数 name に "Python" が渡り、"Hello, " + name の結果 "Hello, Python" が返されます。`return` は関数の戻り値を指定します。

```python
def greet(name):
    return "Hello, " + name

print(greet("Python"))   # Hello, Python
```

### 問47 の解答と解説

**正解： B) 14**

`return a + b` で和 7 が返され、`result` に代入されます。`result * 2` は 14。`return` を書かない関数は `None` を返す点にも注意しましょう。

```python
def add(a, b):
    return a + b   # 戻り値を返す

result = add(3, 4)     # 7
print(result * 2)      # 14
```

### 問48 の解答と解説

**正解： A) 9 / 8**

`exp=2` のようにデフォルト値を指定すると、その引数は省略可能になります。`power(3)` は exp が既定の 2 で 3**2=9、`power(2, 3)` は exp=3 で 2**3=8。デフォルト引数は必ず通常の引数の後ろに置きます。

```python
def power(base, exp=2):   # exp は省略可（既定2）
    return base ** exp

print(power(3))      # 9
print(power(2, 3))   # 8
```

### 問49 の解答と解説

**正解： A) Taro is 20**

`引数名=値` の形（キーワード引数）で渡すと、順番に関係なく名前で対応づけられます。よって `age=20, name="Taro"` でも name="Taro"、age=20 となり "Taro is 20" が返ります。

```python
def introduce(name, age):
    return name + " is " + str(age)

print(introduce(age=20, name="Taro"))   # Taro is 20
```

### 問50 の解答と解説

**正解： D) f(a=1, 2, 3)**

キーワード引数の後ろに位置引数を置くと `SyntaxError` になります。D は `a=1`（キーワード）の後に位置引数 2, 3 が来ているため不正です。位置引数は必ずキーワード引数より前に書きます。

```python
def f(a, b, c):
    return (a, b, c)

print(f(1, c=3, b=2))   # (1, 2, 3)
```

### 問51 の解答と解説

**正解： A) 10**

引数名の前に `*` を付けると、渡された位置引数をタプルとしてまとめて受け取れます（慣習的に `*args`）。ここでは nums=(1,2,3,4) となり `sum(nums)` は 10 です。

```python
def total(*nums):        # 可変長位置引数 → タプル
    return sum(nums)

print(total(1, 2, 3, 4))   # 10
```

### 問52 の解答と解説

**正解： A) {'name': 'Taro', 'age': 20}**

引数名の前に `**` を付けると、キーワード引数を辞書としてまとめて受け取れます（慣習的に `**kwargs`）。ここでは info={'name':'Taro','age':20} になります。

```python
def show(**info):        # 可変長キーワード引数 → 辞書
    return info

print(show(name="Taro", age=20))   # {'name': 'Taro', 'age': 20}
```

### 問53 の解答と解説

**正解： A) 6**

呼び出し時に `*リスト` と書くと、要素を位置引数に展開して渡せます。`add(*[1,2,3])` は `add(1,2,3)` と同じで、和 6 を返します。辞書は `**辞書` でキーワード引数に展開できます。

```python
def add(a, b, c):
    return a + b + c

nums = [1, 2, 3]
print(add(*nums))   # add(1,2,3) と同じ → 6
```

### 問54 の解答と解説

**正解： A) 5 / 10**

関数内で代入された `x` はローカル変数で、外側（グローバル）の x とは別物です。`change()` は 5 を返しますが、グローバルの x は 10 のまま変わりません。関数内の代入は既定でローカルスコープになります。

```python
x = 10
def change():
    x = 5        # これはローカル変数
    return x

print(change())   # 5
print(x)          # 10（グローバルは不変）
```

### 問55 の解答と解説

**正解： A) global**

関数内でグローバル変数に代入したい場合は `global 変数名` を宣言します。これがないと `count += 1` はローカル変数への代入とみなされ `UnboundLocalError` になります。多用は避け、必要な場面に限って使うのが原則です。

```python
count = 0
def increment():
    global count   # グローバル変数を指定
    count += 1

increment()
print(count)   # 1
```

### 問56 の解答と解説

**正解： A) hi / None**

`show("hi")` は "hi" を表示しますが `return` がないため戻り値は `None` です。よって `result` は None になり、1行目に "hi"、2行目に "None" が出力されます。「表示する（print）」ことと「値を返す（return）」ことは別だと押さえましょう。

```python
def show(msg):
    print(msg)     # 表示するだけ（return なし）

result = show("hi")   # hi
print(result)         # None
```

### 問57 の解答と解説

**正解： A) 25**

`lambda 引数: 式` は無名関数（ラムダ式）で、簡単な処理を1行で書けます。`square = lambda x: x ** 2` は `def square(x): return x**2` と同等で、`square(5)` は 25 です。`sorted` の key など一時的な関数によく使われます。

```python
square = lambda x: x ** 2   # 無名関数
print(square(5))   # 25
```

### 問58 の解答と解説

**正解： A) [(2, 'a'), (1, 'b'), (3, 'c')]**

`key=lambda p: p[1]` は各要素の2番目（文字）を並べ替えの基準にします。文字は "a" < "b" < "c" の順なので、(2,'a'),(1,'b'),(3,'c') の順になります。`key` には「要素から比較キーを取り出す関数」を渡します。

```python
pairs = [(1, "b"), (2, "a"), (3, "c")]
result = sorted(pairs, key=lambda p: p[1])   # 2番目の要素で並べ替え
print(result)   # [(2, 'a'), (1, 'b'), (3, 'c')]
```

### 問59 の解答と解説

**正解： A) __doc__**

関数定義の先頭に置いた文字列リテラルは docstring として `関数名.__doc__` に格納されます。`greet.__doc__` は "あいさつを返す関数" を返します。docstring は `help()` でも参照され、関数の説明を書く標準的な方法です。

```python
def greet():
    """あいさつを返す関数"""
    return "hi"

print(greet.__doc__)   # あいさつを返す関数
```

### 問60 の解答と解説

**正解： A) 1 5**

`return a, b` はタプル `(a, b)` を返し、受け取り側で `low, high = ...` と書くとアンパックされて各変数に代入されます。`min` が 1、`max` が 5 なので "1 5" が出力されます。関数は複数の値をまとめて返せます。

```python
def minmax(nums):
    return min(nums), max(nums)   # タプルで返す

low, high = minmax([3, 1, 4, 1, 5])
print(low, high)   # 1 5
```
