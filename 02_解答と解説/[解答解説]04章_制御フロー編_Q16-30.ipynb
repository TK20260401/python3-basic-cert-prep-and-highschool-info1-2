### 問16 の解答と解説

**正解： B) Zero**

`if` が偽なら `elif`（else if）で次の条件を評価し、いずれも偽なら `else` に進みます。x=0 は `x < 0` が偽、`x == 0` が真なので "Zero" が表示されます。Python には switch 文がなく、`if / elif / else` の連鎖で代用します。

```python
x = 0
if x < 0:
    print("Negative")
elif x == 0:   # else if の意味
    print("Zero")
else:
    print("Positive")   # Zero
```

### 問17 の解答と解説

**正解： B) 10**

`range(5)` は 0,1,2,3,4 を生成します（終端の 5 は含みません）。合計は 0+1+2+3+4 = 10。`range(stop)` は 0 から始まり stop の直前まで、という点を押さえましょう。

```python
total = 0
for i in range(5):   # 0,1,2,3,4
    total += i
print(total)   # 10
```

### 問18 の解答と解説

**正解： A) 2 5 8**

`range(start, stop, step)` は start から stop の手前まで step ずつ進みます。2, 5, 8 の次は 11 で stop(10) を超えるため出力されません。`print(..., end=" ")` で改行せず空白区切りにできます。

```python
for i in range(2, 10, 3):   # 2,5,8
    print(i, end=" ")   # 2 5 8
```

### 問19 の解答と解説

**正解： B) 24**

`while` は条件が真の間くり返します。n が 1→2→3→4 と進む間 result に掛け合わされ、1×2×3×4 = 24。条件を `n <= 4` とすると n=4 まで実行されます。カウンタ更新 `n += 1` を忘れると無限ループになる点に注意しましょう。

```python
n = 1
result = 1
while n <= 4:   # n=1,2,3,4 の間くり返す
    result *= n
    n += 1
print(result)   # 24
```

### 問20 の解答と解説

**正解： B) 0 1 2**

`break` はループ自体を即座に終了します。i が 3 になった時点で `print` の前に break されるため、表示されるのは 0, 1, 2 まで。`break` は最も内側のループだけを抜けます。

```python
for i in range(10):
    if i == 3:
        break   # ループを打ち切る
    print(i)   # 0,1,2
```

### 問21 の解答と解説

**正解： B) 1 3 5**

`continue` はそれ以降の処理を飛ばして次の繰り返しに進みます。i が偶数(0,2,4)のときは `print` に到達せずスキップされ、奇数(1,3,5)だけが表示されます。`break` が「ループ終了」なのに対し、`continue` は「今回だけスキップ」です。

```python
for i in range(6):
    if i % 2 == 0:
        continue   # 偶数はスキップ
    print(i, end=" ")   # 1 3 5
```

### 問22 の解答と解説

**正解： B) 0 1 2 done**

ループに付けた `else` は「`break` で中断されずにループが正常終了したとき」に実行されます。ここでは break がないので 0,1,2 を表示後に "done" が出力されます。探索して見つからなかった場合の処理などに使われます。

```python
for i in range(3):
    print(i)   # 0,1,2
else:
    print("done")   # break されなかったので実行される
```

### 問23 の解答と解説

**正解： A) 9 = 3 * 3**

x=3 のとき `9 % 3 == 0` が真になり "9 = 3 * 3" を表示して `break` します。break でループを抜けたため `else` 節は実行されません。この for-else は素数判定の定番パターンで、公式チュートリアルでも紹介されています。

```python
n = 9
for x in range(2, n):
    if n % x == 0:
        print(n, "=", x, "*", n // x)   # 9 = 3 * 3
        break
else:
    print(n, "is a prime number")   # break したので実行されない
```

### 問24 の解答と解説

**正解： C) pass**

`pass` は「何もしない」文で、構文上ブロックが必要だが処理をまだ書かない場合のプレースホルダに使います。空欄のままだとインデントされた本体が無くエラーになります。`continue` や `break` は動作が異なり、ここでは不適切です。

```python
for i in range(3):
    pass   # 何もしない（プレースホルダ）
print("ok")   # ok
```

### 問25 の解答と解説

**正解： A) 0 apple / 1 banana / 2 cherry**

`enumerate()` はイテラブルを回しながら (インデックス, 要素) のペアを返します。既定ではインデックスは 0 から始まります。`enumerate(fruits, start=1)` とすれば 1 から数え始めることもできます。

```python
fruits = ["apple", "banana", "cherry"]
for i, name in enumerate(fruits):
    print(i, name)   # 0 apple / 1 banana / 2 cherry
```

### 問26 の解答と解説

**正解： B) 3**

`while True` は無条件にくり返す書き方で、内部の `if ...: break` で終了条件を作ります。count は 1→2→3 と増え、3 で `count >= 3` が真になり break。ループ後の count は 3 です。終了条件を必ず用意しないと無限ループになる点に注意します。

```python
count = 0
while True:
    count += 1
    if count >= 3:   # 3 以上で抜ける
        break
print(count)   # 3
```

### 問27 の解答と解説

**正解： A) 出力は2行（1 2 3 / 2 4 6）**

外側の `range(1, 3)` は i=1,2 の2回、内側の `range(1, 4)` は j=1,2,3 の3回です。内側ループ終了ごとの `print()` が改行を入れるため、i=1 で "1 2 3"、i=2 で "2 4 6" の計2行になります。

```python
for i in range(1, 3):     # i = 1, 2
    for j in range(1, 4): # j = 1, 2, 3
        print(i * j, end=" ")
    print()   # 1 2 3 / 2 4 6 の2行
```

### 問28 の解答と解説

**正解： A) 1**

`in` 演算子は左辺が右辺（文字列やリストなど）に含まれるかを True / False で返します。"python" の中で母音("aeiou")に該当するのは "o" の1文字だけなので count は 1 です。

```python
vowels = "aeiou"
word = "python"
count = 0
for ch in word:
    if ch in vowels:   # 母音に含まれるか
        count += 1
print(count)   # 1（o のみ）
```

### 問29 の解答と解説

**正解： B) 0 1 2 loop end**

`while` にも `else` を付けられ、条件が偽になってループが正常終了したとき（`break` されなかったとき）に実行されます。0,1,2 を表示後、`n < 3` が偽になり else の "loop end" が出力されます。

```python
n = 0
while n < 3:
    print(n)   # 0,1,2
    n += 1
else:
    print("loop end")   # break されず終了したので実行
```

### 問30 の解答と解説

**正解： A) append / for**

リストの末尾に要素を足すメソッドは `append()` です。同じ処理はリスト内包表記 `[式 for 変数 in イテラブル]` でも簡潔に書けます。どちらも `[0, 1, 4, 9, 16]` を作ります。内包表記は for ループの短縮表現と考えると理解しやすいです。

```python
squares = []
for x in range(5):
    squares.append(x ** 2)
print(squares)   # [0, 1, 4, 9, 16]

squares = [x ** 2 for x in range(5)]
print(squares)   # [0, 1, 4, 9, 16]
```
