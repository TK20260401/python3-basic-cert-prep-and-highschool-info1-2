### 問61 の解答と解説

**正解： A) 4.0**

`import math` でモジュール全体を取り込み、`math.関数名` の形で使います。`math.sqrt(16)` は平方根を返し、結果は浮動小数点数 `4.0` です。`import` はモジュール名前空間ごと読み込む基本の書き方です。

```python
import math
print(math.sqrt(16))   # 4.0
```

### 問62 の解答と解説

**正解： A) 3.14**

`from モジュール import 名前` で、特定の名前を直接（モジュール名を付けずに）使えるようにします。`pi` をそのまま参照でき、`round(pi, 2)` は 3.14 になります。

```python
from math import pi
print(round(pi, 2))   # 3.14
```

### 問63 の解答と解説

**正解： A) 3**

`import モジュール as 別名` で短い名前を付けられます。`m.floor(3.7)` は小数点以下を切り捨てて 3 を返します（`floor` は「以下の最大の整数」）。numpy を np とするなど、慣習的な別名でよく使われます。

```python
import math as m
print(m.floor(3.7))   # 3
```

### 問64 の解答と解説

**正解： A) 4 3**

`from math import ceil, floor` のようにカンマ区切りで複数取り込めます。`ceil(3.2)` は切り上げで 4、`floor(3.8)` は切り捨てで 3。切り上げ・切り捨ての向きに注意しましょう。

```python
from math import ceil, floor
print(ceil(3.2), floor(3.8))   # 4 3
```

### 問65 の解答と解説

**正解： A) 1 以上 6 以下の整数**

`random.randint(a, b)` は a 以上 b 以下（両端を含む）の整数を返します。よって `randint(1, 6)` はサイコロと同じ 1〜6 です。0以上1未満の小数が欲しいときは `random.random()` を使います。

```python
import random
n = random.randint(1, 6)   # 1以上6以下
print(n)
```

### 問66 の解答と解説

**正解： A) リストの中からランダムに1つの要素を返す**

`random.choice(シーケンス)` は、リストなどから要素を1つランダムに選んで返します。全体を並べ替えたい場合は `random.shuffle()`、複数選ぶ場合は `random.sample()` を使います。

```python
import random
fruits = ["apple", "banana", "cherry"]
print(random.choice(fruits))   # ランダムに1つ
```

### 問67 の解答と解説

**正解： A) dir(math) はモジュール内の名前のリストを返すので <class 'list'>**

`dir(モジュール)` は、そのモジュールで定義されている名前（関数・定数など）を文字列のリストで返します。対話的にモジュールの内容を調べるのに便利です。引数なしの `dir()` は現在のスコープの名前を返します。

```python
import math
print(type(dir(math)))   # <class 'list'>
```

### 問68 の解答と解説

**正解： A) ファイルを直接実行したときだけ main() を呼び、import されたときは呼ばない**

モジュールを直接実行すると `__name__` は `"__main__"` になり、他ファイルから import された場合はモジュール名になります。この違いを利用して「直接実行時だけ動かす処理」を書くのが定番パターンです。

```python
def main():
    print("実行されました")

if __name__ == "__main__":
    main()
```

### 問69 の解答と解説

**正解： A) 3**

`sys` モジュールはインタプリタや実行環境に関する情報・機能を提供します。`sys.version_info` はバージョン情報を持つオブジェクトで、`.major` はメジャーバージョン番号（Python 3 系なら 3）です。`sys.argv`（コマンドライン引数）などもよく使われます。

```python
import sys
print(sys.version_info.major)   # 3
```

### 問70 の解答と解説

**正解： A) getcwd() は現在の作業ディレクトリ（カレントディレクトリ）を文字列で返す**

`os` モジュールはOSに依存する機能（ファイル・ディレクトリ操作、環境変数など）を提供します。`os.getcwd()` はカレントディレクトリのパスを文字列で返します。ディレクトリ内容の一覧は `os.listdir()` です。

```python
import os
print(type(os.getcwd()))   # <class 'str'>
```

### 問71 の解答と解説

**正解： A) 2025 が表示される**

`datetime.date(年, 月, 日)` で日付オブジェクトを作れます。`.year` / `.month` / `.day` でそれぞれの値を取り出せるので、`d.year` は 2025 です。現在日時は `datetime.datetime.now()` で取得できます。

```python
import datetime
d = datetime.date(2025, 1, 15)
print(d.year)   # 2025
```

### 問72 の解答と解説

**正解： A) 8.0**

`math.pow(x, y)` は x の y 乗を**浮動小数点数**で返すため、結果は `8.0` です。整数のべき乗が欲しい場合は演算子 `2 ** 3`（結果 8）を使います。同じ「べき乗」でも返る型が異なる点に注意しましょう。

```python
import math
print(math.pow(2, 3))   # 8.0
```

### 問73 の解答と解説

**正解： A) モジュール内の多くの名前を一括で取り込むが、名前の衝突が起きやすく非推奨**

`from module import *` は多数の名前を一度に取り込めますが、どの名前が入ったか分かりにくく、既存の名前と衝突する恐れがあるため一般には推奨されません。必要な名前を明示的に import するのが良い習慣です。

```python
from math import *
print(sqrt(9))   # 3.0
```

### 問74 の解答と解説

**正解： A) os のサブモジュール os.path の join でパスを連結する**

`os.path` は os パッケージのサブモジュールで、パス文字列を扱う関数を提供します。`os.path.join()` はOSに応じた区切り文字でパスを連結します（環境により "dir/file.txt" など）。単純な文字列連結より安全です。

```python
import os
print(os.path.join("dir", "file.txt"))   # 例: dir/file.txt
```

### 問75 の解答と解説

**正解： A) 5.0**

`from math import sqrt as root` は、取り込む名前 `sqrt` に `root` という別名を付けます。以降 `root(25)` として使え、平方根 `5.0`（浮動小数点数）を返します。`as` は import と from-import の両方で別名付けに使えます。

```python
from math import sqrt as root
print(root(25))   # 5.0
```
