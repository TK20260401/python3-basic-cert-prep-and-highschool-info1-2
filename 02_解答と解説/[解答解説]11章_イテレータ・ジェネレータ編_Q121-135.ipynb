### 問121 解答
**正解: B. `iter`**

`iter(オブジェクト)` は反復可能オブジェクト(iterable)からイテレータ(iterator)を取得する組み込み関数です。取得したイテレータは `list_iterator` などの型になります。

```python
nums = [10, 20, 30]
it = iter(nums)
print(type(it).__name__)
```

### 問122 解答
**正解: C. `next`**

`next(イテレータ)` はイテレータから次の要素を1つ取り出します。呼び出すたびに次の要素へ進み、要素が尽きると `StopIteration` 例外が送出されます。

```python
it = iter([1, 2, 3])
print(next(it))
print(next(it))
```

### 問123 解答
**正解: B. `StopIteration`**

イテレータの要素をすべて取り出した後にさらに `next()` を呼ぶと `StopIteration` が送出されます。`for` 文はこの例外を内部で捕捉してループを終了しています。

```python
it = iter([1])
next(it)
try:
    next(it)
except StopIteration:
    print("終了")
```

### 問124 解答
**正解: B. `yield`**

`yield` を含む関数はジェネレータ関数になります。`yield` に到達すると値を1つ返して実行を一時停止し、次に呼ばれると続きから再開します。

```python
def gen():
    yield 1
    yield 2

for v in gen():
    print(v)
```

### 問125 解答
**正解: B. `( )`(丸かっこ)**

丸かっこで囲むとジェネレータ式になり、全要素を一度にメモリ展開せず遅延評価します。角かっこならリスト内包表記になります。ここでは `0+1+4+9+16=30` が出力されます。

```python
gen = (x * x for x in range(5))
print(sum(gen))
```

### 問126 解答
**正解: B. `enumerate`**

`enumerate(iterable)` は `(インデックス, 要素)` のタプルを順に生成します。`for i, x in enumerate(...)` の形でインデックスと要素を同時に受け取れます。

```python
for i, name in enumerate(["a", "b", "c"]):
    print(i, name)
```

### 問127 解答
**正解: A. `start=1`**

`enumerate(iterable, start=1)` とすると番号を1から開始できます。出力は `1 春` / `2 夏` となります。

```python
for i, name in enumerate(["春", "夏"], start=1):
    print(i, name)
```

### 問128 解答
**正解: C. `zip`**

`zip(iter1, iter2, ...)` は各iterableの同じ位置の要素をタプルにまとめて生成します。最も短いiterableに長さが合わせられます。

```python
names = ["A", "B"]
scores = [90, 80]
for n, s in zip(names, scores):
    print(n, s)
```

### 問129 解答
**正解: A. `next(it, "なし")`**

`next(イテレータ, 既定値)` と第2引数を渡すと、要素が尽きたときに `StopIteration` を送出せず既定値を返します。この例では `"なし"` が出力されます。

```python
it = iter([])
print(next(it, "なし"))
```

### 問130 解答
**正解: B. `__next__`**

イテレータプロトコルでは `__iter__`(自身を返す)と `__next__`(次の要素を返す)を実装します。要素が尽きたら `__next__` 内で `StopIteration` を送出します。

```python
class Count:
    def __init__(self):
        self.n = 0
    def __iter__(self):
        return self
    def __next__(self):
        self.n += 1
        if self.n > 2:
            raise StopIteration
        return self.n

for v in Count():
    print(v)
```

### 問131 解答
**正解: B. `yield from`**

`yield from iterable` は、そのiterableの各要素を順に `yield` するのと同じ動作をします。ネストしたループを書かずに委譲できます。出力は `[1, 2, 3]` です。

```python
def gen():
    yield from [1, 2, 3]

print(list(gen()))
```

### 問132 解答
**正解: B. `list`**

`list(ジェネレータ)` はジェネレータを最後まで反復して全要素をリストに集めます。出力は `[1, 2, 3]` です。ジェネレータは一度消費すると再利用できない点に注意します。

```python
gen = (x + 1 for x in range(3))
result = list(gen)
print(result)
```

### 問133 解答
**正解: B. 短い方に合わせる**

`zip` は最も短いiterableに長さを合わせて停止します。ここでは `b` の長さ2に合わせられ、`list(zip(a, b))` は要素2個となり出力は `2` です。

```python
a = [1, 2, 3]
b = ["x", "y"]
print(len(list(zip(a, b))))
```

### 問134 解答
**正解: A. `sum(i for i in range(4))`**

関数呼び出しの唯一の引数がジェネレータ式のとき、外側の丸かっこを兼ねられるため `sum(i for i in range(4))` と書けます。`0+1+2+3=6` が出力されます。

```python
total = sum(i for i in range(4))
print(total)
```

### 問135 解答
**正解: B. `reversed`**

`reversed(シーケンス)` は要素を逆順にたどるイテレータを返します。出力は `c` / `b` / `a` の順です。なおリストの `reverse()` メソッドは元のリストを破壊的に反転する点で異なります。

```python
for c in reversed("abc"):
    print(c)
```
