### 問76 解答：A. `format`
`str.format()` は文字列内の `{}`（置換フィールド）を引数で順に置き換えます。位置引数は `{}` に前から順に対応します。

```python
name = 'Alice'
age = 30
print('{} さん ({} 歳)'.format(name, age))
```

### 問77 解答：C. `f`
接頭辞 `f` を付けると f-string になり、`{式}` の中に変数や式を直接書けます。実行時に評価されて文字列へ埋め込まれます。

```python
name = 'Bob'
score = 95
print(f'{name} の得点は {score} 点')
```

### 問78 解答：B. `:.2f`
f-string の `{値:.2f}` は「小数点以下2桁の浮動小数点数」を意味します。`.2` が精度、`f` が固定小数点表記です。

```python
import math
print(f'{math.pi:.2f}')
```

### 問79 解答：C. `>5`
`>` は右寄せ、`<` は左寄せ、`^` は中央寄せを表します。`>5` は「幅5で右寄せ」となり、不足分は空白で埋められます。

```python
print('{:>5}'.format(42))
```

### 問80 解答：B. `sep='-'`
`print()` の `sep` は複数の引数の間に挿入される区切り文字（既定は空白）です。`sep='-'` とすると各値が `-` で連結されます。

```python
print('2026', '04', '01', sep='-')
```

### 問81 解答：B. `end=''`
`print()` の `end` は出力の末尾に付ける文字列（既定は改行 `\n`）です。`end=''` にすると改行されず、次の出力が同じ行に続きます。

```python
print('A', end='')
print('B')
```

### 問82 解答：C. `int`
`input()` は入力を必ず文字列で返します。数値計算に使うには `int()` などで型変換（キャスト）が必要です。

```python
s = input('数を入力: ')
n = int(s)
print(n + 1)
```

### 問83 解答：B. `'r'`
`'r'` は読み込み（read）モードで既定値です。`'w'` は書き込み（上書き）、`'a'` は追記、`'x'` は新規作成専用です。

```python
f = open('data.txt', 'r')
text = f.read()
f.close()
```

### 問84 解答：C. `write`
ファイルオブジェクトの `write()` は文字列をファイルに書き込みます。`with` 文を使うとブロックを抜けるときに自動で `close()` されます。

```python
with open('out.txt', 'w') as f:
    f.write('Hello\n')
```

### 問85 解答：B. 1行ずつ
ファイルオブジェクトはイテラブルで、`for` で回すと1行ずつ（改行文字を含む文字列）取り出せます。`rstrip()` で末尾の改行を除去しています。

```python
with open('data.txt') as f:
    for line in f:
        print(line.rstrip())
```

### 問86 解答：A. `strip`
`strip()` は文字列の先頭と末尾の空白（や指定文字）を取り除きます。片側だけなら `lstrip()`／`rstrip()` を使います。

```python
s = '  hello  '
print(s.strip())
```

### 問87 解答：B. `split`
`split(sep)` は区切り文字 `sep` で文字列を分割し、リストを返します。引数を省略すると空白で分割します。

```python
s = 'a,b,c'
print(s.split(','))
```

### 問88 解答：C. `join`
`区切り文字.join(反復可能)` は各要素を区切り文字で連結した文字列を返します。要素はすべて文字列である必要があります。

```python
parts = ['a', 'b', 'c']
print('-'.join(parts))
```

### 問89 解答：A. `replace`
`replace(old, new)` は文字列中のすべての `old` を `new` に置き換えた新しい文字列を返します（元の文字列は変更されません）。

```python
s = '2026/04/01'
print(s.replace('/', '-'))
```

### 問90 解答：B. `!r`
f-string の `{値!r}` は `repr()` を適用してから埋め込みます。文字列なら引用符付きで `'Bob'` のように表示されます（`!s` は str、`!a` は ascii）。

```python
name = 'Bob'
print(f'{name!r}')
```
